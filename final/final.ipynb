## Imports and Functions

````python
# Imports only
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Functions only

def find_differences(ref_df, compare_df):
    # Find differences in values
    differences = ref_df.ne(compare_df)

    # Display the differences
    print("Differences in Values:")
    return differences

def preprocess_datasets(df):
    # Fill NaN values with None
    df.fillna(value= np.NaN, inplace=True)

    # Remove duplicates
    df.drop_duplicates(inplace=True)
    return df
````

## Analyzing What Influences Most

````python
plt.figure(figsize=[15, 8])
df[df['treatment'] == 'Yes']['Age'].plot.kde(color='blue')
df[df['treatment'] == 'No']['Age'].plot.kde(color='orangered')
plt.xticks(ticks=np.arange(0, 100, 5), size=12)
plt.xlabel(xlabel='Age', size=14)
plt.ylabel(ylabel='Density', size=14)
plt.title(label='Density of Age Feature', size=16)
plt.show()

plt.figure(figsize=[20, 8])
sns.countplot(x='Age', hue='treatment', data= df)
plt.title(label='Age vs Treatment', size=16)
plt.xlabel(xlabel='Age', size=14)
plt.ylabel(ylabel='Frequency', size=14)
plt.show()
````

**Insights:**

Individuals aged 30 and above appear to be more conscious and proactive about addressing their mental health concerns, as evidenced by a higher proportion seeking mental health treatment compared to younger age groups. This insight underscores the importance of targeted mental health interventions and support tailored to different age demographics within the surveyed population.

````python
figure = plt.figure(figsize=[20, 20])

plt.subplot(2,2,1)
df['treatment'][df['Gender'] == 'Male'].value_counts().plot(kind='pie', autopct='%1.1f%%', wedgeprops = dict(width = 0.15), startangle=90)
plt.title(label='Treatment in Males', size=16)

plt.subplot(2,2,2)
df['treatment'][df['Gender'] == 'Female'].value_counts().plot.pie(autopct='%1.1f%%',wedgeprops = dict(width = 0.15), startangle=90)
plt.title(label='Treatment in Females', size=16)

plt.subplot(2,2,3)
df['treatment'][df['Gender'] == 'LGBTQIA+'].value_counts().plot.pie(autopct='%1.1f%%', wedgeprops = dict(width = 0.15), startangle=90)
plt.title(label='Treatment in LGBTQIA+', size=16)

plt.show()
````

**Insights:**

Based on the graphs, there is a notable difference in the propensity to seek mental health treatment across different gender identities. Females and individuals identifying as LGBTQIA+ show a higher likelihood of seeking treatment compared to males. This insight emphasizes the importance of considering diverse gender identities in mental health awareness campaigns and support initiatives.

````python
# Create a cross-tabulation of Treatment and Work Interference
cross_tab = pd.crosstab(df['treatment'], df['work_interfere'])

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))
sns.countplot(x='work_interfere', hue='treatment', data=df, order=df['work_interfere'].value_counts().index)
plt.title('Association between Treatment and Work Interference', fontsize=16)
plt.xlabel('Work Interference', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Treatment', loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.show()
````

**Insights:**

Employees who experience interference "Often" or "Rarely" are more inclined to seek treatment, suggesting a potential correlation between work-related challenges and mental health concerns. This insight emphasizes the importance of addressing work-related stressors and providing support for employees in managing their mental well-being.

````python
plt.figure(figsize=[15, 5])
plt.subplot(1, 2, 1)
df.groupby(['treatment', 'family_history'])['family_history'].count()['No'].plot.bar(color='lightcoral')
plt.xlabel(xlabel='Family_history - No', size=14)
plt.ylabel(ylabel='Undergone Treatment', size=14)
plt.title(label='Treatment vs Family History (No Family History)', size=16)

plt.subplot(1, 2, 2)
df.groupby(['treatment', 'family_history'])['family_history'].count()['Yes'].plot.bar(color='seagreen')
plt.xlabel(xlabel='Family_history - Yes', size=14)
plt.ylabel(ylabel='Undergone Treatment', size=14)
plt.title(label='Treatment vs Family History (With Family History)', size=16)

plt.tight_layout()
plt.show()
````

**Insights:**

Employees with a family history of mental health issues demonstrate a higher propensity to seek treatment, indicating a potential correlation between family history and the likelihood of addressing mental health concerns. Conversely, individuals without a family history may exhibit lower awareness or a reduced inclination to pursue mental health treatment.

This observation underscores the importance of considering family history as a factor in understanding and addressing mental health challenges in the workplace.

````python
# Filter data for individuals seeking treatment ('Yes')
df_yes = df[df['treatment'] == 'Yes']

# Get the top 10 countries with the highest number of individuals seeking treatment
top_countries_yes = df_yes['Country'].value_counts().head(10)

# Calculate the percentage of individuals seeking treatment in each country among the top 10
percentage_yes_top_countries = (top_countries_yes / df['Country'].value_counts().head(5)).sort_values(ascending=False)

# Display the results, dropping NaN values
result_df = percentage_yes_top_countries.dropna()

plt.figure(figsize=[15,8])
result_df.dropna().sort_values(ascending=False).plot.bar(color='navy')

plt.xticks(rotation='vertical', size=14)
plt.xlabel(xlabel ='Top 5 Countries', size=14)
plt.ylabel(ylabel ='Frequency', size=14)
plt.title(label ='Top 5 countries - Actually focussing on Treating Mental Health issues', size=16)

plt.show()
````

**Insights:**

United States and Canada. lead in prioritizing the resolution of employees' mental health issues, encouraging a higher number of individuals to seek treatment, followed by the UK and Ireland

- Strategies to mitigate work interference should be explored to create a conducive work environment that prioritizes employee well-being.
- Large organizations should implement tailored mental health programs to address the unique challenges faced by employees in diverse workplace settings.
